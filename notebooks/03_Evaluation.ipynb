# Evaluation
This notebook computes final metrics, confusion matrix, and ROC curves.

```python
from pathlib import Path
import sys
import yaml
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'configs').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

sys.path.append(str(PROJECT_ROOT))

from src.evaluate import evaluate_model

config_path = PROJECT_ROOT / 'configs' / 'config.yaml'
with open(config_path, 'r', encoding='utf-8') as file:
    config = yaml.safe_load(file)

checkpoint_path = PROJECT_ROOT / 'checkpoints' / 'best_model.pth'
```

```python
metrics = evaluate_model(config, checkpoint_path)
metrics
```

```python
from matplotlib import image as mpimg

figures_dir = PROJECT_ROOT / config['paths']['figures_dir']

confusion_path = figures_dir / 'confusion_matrix.png'
roc_path = figures_dir / 'roc_curve.png'

if confusion_path.exists():
    plt.figure(figsize=(6, 5))
    plt.imshow(mpimg.imread(confusion_path))
    plt.axis('off')
    plt.title('Confusion Matrix')
    plt.show()

if roc_path.exists():
    plt.figure(figsize=(7, 5))
    plt.imshow(mpimg.imread(roc_path))
    plt.axis('off')
    plt.title('ROC Curve')
    plt.show()
```
